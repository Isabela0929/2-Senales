```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def kuramoto_osc1p(t,x):
 rho,theta= x
 dx = [rho *(1-rho**2), 1]
 return dx

kuramoto_sol = solve_ivp (kuramoto_osc1p, [0,2*np.pi], [1,0],dense_output=True)
print(kuramoto_sol)

plt.plot(kuramoto_sol.t,kuramoto_sol.y[0])
plt.scatter(kuramoto_sol.t,kuramoto_sol.y[1])
plt.show()

```

```python
tplot= np.linspace(0,2*np.pi,100)
rho= kuramoto_sol.sol(tplot)[0]
theta= kuramoto_sol.sol(tplot)[1]
print(tplot)

xt=rho*np.cos(theta)
yt=rho*np.sin(theta)
plt.plot(xt,yt)
#print(xt)
#print(yt)
plt.show()
```

```python
for rho0 in np.linspace(0,20,10):
  for thetha0 in np.linspace(0,3*np.pi/2.4):
   kuramoto_sol= solve_ivp(kuramoto_osc1p,[0,np.pi*2],[rho0,thetha0], dense_output=True)
   rho= kuramoto_sol.sol(tplot)[0]
   theta= kuramoto_sol.sol(tplot)[1]
   xt=kuramoto_sol.y[0]*np.cos(kuramoto_sol.y[1])
   yt=kuramoto_sol.y[0]*np.sin(kuramoto_sol.y[1])
   plt.plot(xt,yt)
plt.show()
```
